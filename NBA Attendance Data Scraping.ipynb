```python
import urllib
import urllib.request
from bs4 import BeautifulSoup
import os
import string
```

```python
def make_soup(url):
    thepage = urllib.request.urlopen(url)
    soupdata = BeautifulSoup(thepage, "html.parser")
    return soupdata
```

```python
#url to be scraped
url = 'https://www.basketball-reference.com/leagues/NBA_2017_games-october.html'

# html from the url
soup = make_soup(url)
    

```

```python
#script to scrape the data
years =('2013','2014','2015','2016','2017')
months = ('october', 'november','december','january','february','march','april','may','june')
gamedatasaved=""
for year in years:
    for month in months:
        try:
            soup = make_soup("https://www.basketball-reference.com/leagues/NBA_"+year+"_games-"+ month +".html")
        except urllib.error.HTTPError:
                print('{} not found'.format(url))
                continue
            
        for record in soup.findAll('tr'):
            gamedata=''
            for date in record.findAll('th',scope = 'row'):
                gamedata = gamedata + date.text
                for data in record.findAll('td'):
                    #print(data.text)
                    gamedata =  gamedata + ","+ data.text
                gamedatasaved = gamedatasaved +"\n" + gamedata
print(gamedatasaved)
        
```

```python
#Write scraped data to a csv file
header="Weekday,Date,Year,Time,Visitor,V_PTS,Home,H_PTS,Box Score,OT,Attend, ,F_City, F_Country"
file = open(os.path.expanduser('GameData.csv'),'wb')
file.write(bytes(header, encoding='ascii',errors='ignore'))
file.write(bytes(gamedatasaved,encoding='ascii', errors='ignore'))
```
